```python
import pandas as pd

# Creating a DataFrame
df = pd.DataFrame({"Name": ["  Mayur ", "Vaibhav  ", "Sadhana  "],
                   "Gender": ['Male ', 'Male', '  Female']})
print("Original DataFrame:")
print(df)

# Removing right-sided whitespaces from the 'Name' column
trail_df = df['Name'].apply(lambda x: x.rstrip())
print("\nAfter removing right-sided whitespaces (rstrip) from 'Name':")
print(trail_df)

# If you want to handle left-sided or full strip as well:
left_trimmed = df['Name'].apply(lambda x: x.lstrip())  # Removes left-side whitespaces
stripped = df['Name'].apply(lambda x: x.strip())       # Removes both sides' whitespaces

print("\nAfter removing left-sided whitespaces (lstrip) from 'Name':")
print(left_trimmed)

print("\nAfter removing all surrounding whitespaces (strip) from 'Name':")
print(stripped)
```

```python
# extract years between 1800 to 2200 from the specified column of a given DataFrame. (DF given)

yr_df = pd.DataFrame({'company_code': ['c0001','c0002','c0003', 'c0003', 'c0004'],
                   'year': ['year 1800','year 1700','year 2300', 'year 1900', 'year 2200']})
yr_df
yr_df.info()
```

```python
yr_df['year'] = yr_df['year'].apply(lambda x : x.replace('year', ''))
yr_df['year'] = yr_df['year'].astype(int)
yr_df.info()
```

```python
yr_df
```

```python
 # Extract rows where 'year' is between 1800 and 2200
filtered_yr_df = yr_df[yr_df['year'].between(1800, 2200)].reset_index(drop=True)
filtered_yr_df
```

## Same perform it by regex

```python
import pandas as pd
import re

# Example DataFrame
yr_df = pd.DataFrame({'company_code': ['c0001', 'c0002', 'c0003', 'c0003', 'c0004'],
                      'year': ['year 1800', 'year 1700', 'year 2300', 'year 1900', 'year 2200']})

# Define a function to extract years using regex
def extract_year(text):
    match = re.search(r'\b(18[0-9]{2}|19[0-9]{2}|20[0-9]{2}|21[0-9]{2}|2200)\b', text)
    return match.group(0) if match else None

# Apply the function to the 'year' column
yr_df['extracted_year'] = yr_df['year'].apply(extract_year)

# Filter rows where extracted_year is between 1800 and 2200
filtered_yr_df = yr_df.dropna(subset=['extracted_year']).reset_index(drop=True)

print(filtered_yr_df)
```

```python
import pandas as pd

# Example DataFrame
data = {"column": ["In 1999, something happened", "Year 1805 was memorable", 
                   "This happened in 2201", "The year 2000 was special", "1800 and beyond"]}
df = pd.DataFrame(data)
print("Original DataFrame:")
print(df)

# Extracting years between 1800 and 2200 using regex
df['year'] = df['column'].str.extract(r'\b(18[0-9]{2}|19[0-9]{2}|20[0-9]{2}|21[0-9]{2}|2200)\b')

# Print the updated DataFrame
print("\nUpdated DataFrame with extracted years:")
print(df)
```
